## 1 (1,25p). Càrrega dels datasets

Defineix la funció *load_dataset()*, que rebi com argument un array de les rutes de fitxers que contenen els datasets, i retorni un dataframe que contingui totes les dades dels diferents fitxers.

Crea l'array de fitxers. Fes-ho de manera programàtica. És a dir, que serveixi per crear un array que tant podrien ser 4 fitxers csv com 100 fitxers csv. Utilitza rutes relatives.

```python
# TODO
def load_dataset(fitxers) -> pd.DataFrame:
  # definir variable
  df = None
  # recòrrer els fitxers
  for fitxer in fitxers:
    # validar fitxer existeix
    if not os.path.isfile(fitxer):
      # crea exepció per notificar que falta el fitxer
      raise Exception(f"{fitxer} no existeix a la ruta")

    # validar si es primera execució
    if df is None:
      # llegir csv amb ruta fitxer i separar columnes per una ','. Font llibre asignatura
      df = pd.read_csv(fitxer, delimiter=',')
    # barreja el pandas actual amb el llegit font: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html
    else:
      df.merge(pd.read_csv(fitxer, delimiter=','))
  return df

df = load_dataset(["data/2023q1.csv","data/2023q2.csv","data/2023q3.csv","data/2023q4.csv"])
```

## 5. (1p) Aproximació polinòmica

Teniu un histograma de com es distribueixen les denúncies per franges horàries. Us plantegeu el repte de si podeu aproximar aquest comportament amb una funció polinòmica que segueixi la corba de l'histograma. Com més gran sigui el grau del polinomi l'aproximació serà més fidel. Provareu un polinomi de grau 4 i un polinomi de grau 8. Fareu servir *polyfit* (*polynomial fitting*) de NumPy.

Com a resultat final heu d'obtenir una gràfica on es supoerposin els punts de l'histograma, i les corbes dels polinomis de grau 4 i grau 8. Se't proporciona una ajuda de com fer l'aproximació polinòmica d'una funció sinums amb un polinomi de grau 3.

```python
# Ajuda: aproximació polinòmica de la funció sin(x) amb un polinomi de grau 3
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as pol
x = np.linspace(0, 6, 100)
y = np.sin(x)
pol_grau3 = pol.polyfit(x, y, 3)
ypol = pol_grau3[3]*x**3 + pol_grau3[2]*x**2 + pol_grau3[1]*x + pol_grau3[0]

plt.plot(x, y, x, ypol)
```
