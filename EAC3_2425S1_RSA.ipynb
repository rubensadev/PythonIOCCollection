## 2. (2p) Arquitectura bàsica d'xarxa neuronal

Per un conjunt de dades com el dels pingüins, on es tracta d'un problema de classificació multiclasse (3 espècies de pingüins), una bona arquitectura de xarxa neuronal a provar és una xarxa neuronal feedforward, també coneguda com a Multilayer Perceptron (MLP).

Específicament, una arquitectura MLP que pot funcionar bé per aquest dataset és la següent:

* Capa d'entrada: Amb *input_size* que correspondria al nombre de característiques del dataset.

* Dues capes ocultes: Aquestes capes aprenen representacions més complexes de les dades d'entrada. Com que és un problema relativament senzill les farem de només 10 neurones (*hidden_size*).

* Capa de sortida: Amb *output_size* que correspondria al nombre de classes (3 espècies de pingüins).

* Funció d'activació ReLU: S'utilitza entre les capes ocultes per introduir no-linealitat al model.

```python
# Definiu la xarxa
class MLP(nn.Module):
  def __init__(self, input_size = 6, hidden_size = 10, output_size = 3):
    super().__init__()
    self.fc1 = nn.Linear(input_size, hidden_size)
    self.fc2 = nn.Linear(hidden_size, hidden_size)
    self.fc3 = nn.Linear(hidden_size, output_size)

  # funció per evaluar les dades d'entrada i obtenir un resultat
  def forward(self, x):
    x = torch.relu(self.fc1(x))
    x = torch.relu(self.fc2(x))
    # l'ultima capa no fem cap funció d'activació, ja que farem servir cross entropy loss per evaluar la perdua del model a l'entrenament
    x = self.fc3(x)
    return x
```

```python
# Inicialitzeu la xarxa i moveu-la a la GPU si està disponible
# font: https://pytorch.org/docs/stable/generated/torch.nn.Module.html#torch.nn.Module.to
# als apunts s'ha vist com s'inicialitza una Xarxa neuronal amb capes ocultes que s'apliquen una funció per evitar la linealitat
model = MLP().to(device)
```

## 3. (2p) Entrenament

Implementeu el bucle bàsic d'entrenament i feu un gràfic de l'evolució de les pèrdues durant l'entrenament.

```python
# Definiu la funció de pèrdua i l'optimitzador
# Faig servir Adam perque als apunts esmenta que es el mes popular per a la majoría de problemes daprenentatge profund
optimizer = optim.Adam(model.parameters(), lr=0.001)

# escollit CrossEntropyLoss perque es la que millor per a classificació multiple, com hem de predir 3 especies en el nostre cas
criterion = nn.CrossEntropyLoss()
```

```python
# Definiu i executeu el bucle d'entrenament guardant les pèrdues en una llista
num_epochs = 100
loss_list = []
x_train_len = len(X_train)

# Create a TensorDataset from X_train and y_train
train_dataset = TensorDataset(X_train, y_train)

# Create a DataLoader to handle batching and shuffling
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

for epoch in range(num_epochs):
  loss_sum = 0
  # iterar per les dades d'entrenament
  for i, data in enumerate(train_loader, 0):
    # reset a zero el gradient del optimitzador
    optimizer.zero_grad()
    # evaluar les dades d'entrada
    outputs = model(data[0])

    # s'evalua la perdua sense aplicar cap formula de correció perque fem servir cross entropy loss
    loss = criterion(outputs, data[1])
    loss.backward()
    # actualitza els gradients del model
    optimizer.step()
    loss_sum += loss.item()
  if (epoch+1) % 10 == 1:
    print(f"Train epoch {epoch} with loss avg {loss_sum/x_train_len}")
  loss_list.append(loss_sum/x_train_len)
```

```python
# Gràfic de la pèrdua d'entrenament
plt.figure(figsize=(10, 6))
plt.xlabel('Època')
plt.ylabel('Pèrdua total')
plt.plot(loss_list, label="CrossEntropyLoss - Adam")
plt.legend()
plt.grid(True)
plt.show()
# al diagrama es pot apreciar com baixa progressivament
```
